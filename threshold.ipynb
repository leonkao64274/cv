```python
import matplotlib.pyplot as plt
import cv2
```

<font size=6> Read RGB show gray

```python
def read_RGB_show_gray(name="Figure3.jpg"):
    img = cv2.imread(name)
    grayimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return grayimg
grayimg=read_RGB_show_gray()
plt.imshow(grayimg,cmap="gray")
```

<font size=6> Histogram

```python
def make_histogram(img=grayimg):
    histogram=[0]*260
    x=[i for i in range(260)]
    for i in range(len(grayimg)):
        for j in range(len(grayimg[0])):
            scale=grayimg[i][j]
            histogram[scale]+=1
    plt.bar(x,histogram)
    return histogram
histogram=make_histogram()
```

```python
sorted_histo=sorted(histogram)
sorted_histo[-3:]
```

<font size=6> Threshhold

```python
def threshhold(img,n):
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j]>n:
                img[i][j]=255
            else:
                img[i][j]=0
    return img
```

```python
grayimg=read_RGB_show_gray()
img=threshhold(img=grayimg,n=200)
plt.imshow(img,cmap="gray")
```
