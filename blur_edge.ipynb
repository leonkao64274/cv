```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from tqdm import tqdm
```

<font size=6> Color sobel gradient 

```python
def read_RGB_show_gray(name="Block.jpg"):
    img = cv2.imread(name)
    grayimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return grayimg
grayimg=read_RGB_show_gray("Eye.png")
```

```python
# 參數 1,0 爲只在 x 方向求一 導數 最大可以求 2 導數。
sobelx = cv2.Sobel(grayimg, cv2.CV_64F, 1, 0, ksize=5)
# 參數 0,1 爲只在 y 方向求一 導數 最大可以求 2 導數。
sobely = cv2.Sobel(grayimg, cv2.CV_64F, 0, 1, ksize=5)
```

```python
g=np.zeros((grayimg.shape[0],grayimg.shape[1]))
for i in tqdm(range(len(sobelx))):
    for j in range(len(sobelx[0])):
        g[i][j]=(sobelx[i][j]**2+sobely[i][j]**2)**(1/2)
        if g[i][j]>255:
            g[i][j]=255
```

```python
merged = cv2.merge([sobelx,sobely,g])
```

```python
plt.imshow(merged)
```

<font size=6> Gaussian filter

```python
imgName = "pill.jpg"
kernel_size = (85, 85)
#高斯遮罩大小只能奇數乘奇數
img = cv2.imread(imgName)
img1 = cv2.GaussianBlur(img, kernel_size, 1)#1是sigma值，下面依此類推
img2 = cv2.GaussianBlur(img, kernel_size, 2)
img5 = cv2.GaussianBlur(img, kernel_size, 5)
```
