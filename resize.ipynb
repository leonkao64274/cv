```python
import matplotlib.pyplot as plt
import cv2
```

<font size=6> Read and show

```python
def read_RGB(name="Figure2.png"):
    img = cv2.imread(name)
    return img
img=read_RGB()
print(img.shape)
plt.imshow(img)
```

<font size=6> Resize

```python
img2=cv2.resize(img, (158, 150))
plt.imshow(img2)
```

```python
img2=cv2.resize(img, (79, 58))
plt.imshow(img2)
```
