```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from tqdm import tqdm
```

<font size=6> Mean filter

```python
img = cv2.imread("Figure2.png")
blur3 = cv2.blur(img,(3,3))
blur11 = cv2.blur(img,(11,11))
blur19 = cv2.blur(img,(19,19))
```

```python
blurred = np.hstack([blur3,blur11,blur19])
```

```python
plt.imshow(blurred)
```

<font size=6> Median filter

```python
median3=cv2.medianBlur(img,3)
median11=cv2.medianBlur(img,11)
median19=cv2.medianBlur(img,19)
```

```python
medians= np.hstack([
    cv2.medianBlur(img,3),
    cv2.medianBlur(img,11),
    cv2.medianBlur(img,19)])
```

```python
plt.imshow(medians)
```
