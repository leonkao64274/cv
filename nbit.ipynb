```python
import matplotlib.pyplot as plt
import cv2
```

<font size=6> Read, convert RGB to gray and show

```python
def read_RGB_show_gray(name="Block.jpg"):
    img = cv2.imread(name)
    grayimg=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return grayimg
grayimg=read_RGB_show_gray()
plt.imshow(grayimg,cmap="gray")
```

<font size=6> Change scale

```python
def change_scale(grayimg,scale):
    part=scale-1
    block=[round(256/part*i) for i in range(part+1)]
#    印出除了0以外的block+1 
#    print("scale:",block)
    for i in range(len(grayimg)):
            for j in range(len(grayimg[0])):
                if grayimg[i][j]==0:  #必要
                    pass
                elif grayimg[i][j]==255: #太多255 加快速度用
                    pass
                else:
                    for k in range(len(block)-1):
                        if block[k]<grayimg[i][j]+1 and grayimg[i][j]+1<=block[k+1]:
                            grayimg[i][j]=block[k+1]-1

    return grayimg
```

```python
def differient_scale(scales):
    grayimg=read_RGB_show_gray()
    grayimg=change_scale(grayimg=grayimg,scale=scales)
    plt.imshow(grayimg,cmap="gray")
    return grayimg
```

```python
scale8=differient_scale(scales=8)
```

<font size=6> Scales

```python
def scales(img):
    tmp=img.tolist()
    content=[list(set(tmp[i]))[0] for i in range(len(tmp))]
    scales=set(content)
    return scales
scales(scale3)
```
